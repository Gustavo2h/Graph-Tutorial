## Seção 1: Conceitos Básicos e Configuração do Ambiente

Nesta seção, vamos importar as bibliotecas necessárias e construir o nosso primeiro grafo simples.

### O que é um grafo?

Um **grafo** é uma estrutura composta por **nós** (ou vértices) e **arestas** que conectam pares de nós. Essa estrutura é utilizada para modelar relações e interações em diversas áreas, como redes sociais, transporte, biologia, entre outras.

```python
# Importando as bibliotecas necessárias
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

# Configurando parâmetros de plotagem
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 12
```

## Seção 2: Criando e Visualizando um Grafo Simples

Nesta seção, vamos criar um grafo simples com alguns nós e arestas, e visualizar sua estrutura utilizando o NetworkX.

### Exemplo Prático:

Criaremos um grafo não direcionado representando uma pequena rede social, onde cada nó representa uma pessoa e cada aresta representa uma conexão (amizade) entre elas.

```python
# Criando o grafo
G = nx.Graph()

# Adicionando nós (pessoas)
pessoas = ['Ana', 'Bruno', 'Carlos', 'Diana', 'Eduardo']
G.add_nodes_from(pessoas)

# Adicionando arestas (conexões entre as pessoas)
conexoes = [('Ana', 'Bruno'), ('Ana', 'Carlos'), ('Bruno', 'Diana'), ('Carlos', 'Diana'), ('Diana', 'Eduardo')]
G.add_edges_from(conexoes)

# Visualizando o grafo
pos = nx.spring_layout(G)  # Define a posição dos nós
nx.draw(G, pos, with_labels=True, node_color='skyblue', edge_color='gray', node_size=1500, font_size=12)
plt.title('Rede Social Simples')
plt.show()
```

## Seção 3: Explorando Propriedades do Grafo

Agora que criamos o nosso grafo, vamos explorar algumas de suas propriedades, como o número de nós, arestas e o grau (número de conexões) de cada nó.

### Exemplo Prático:

Exiba informações básicas sobre o grafo e imprima o grau de cada nó.

```python
# Informações básicas sobre o grafo
print('Número de nós:', G.number_of_nodes())
print('Número de arestas:', G.number_of_edges())

# Grau de cada nó
print('\nGrau de cada nó:')
for node, degree in G.degree():
    print(f'{node}: {degree}')
```

## Seção 4: Exercícios Práticos

Chegou a sua vez de praticar! Tente resolver os seguintes desafios:

2. **Visualização Customizada:** Experimente alterar os parâmetros de visualização (como cores, tamanhos e layout) para deixar o grafo mais informativo e visualmente agradável.

Após realizar as modificações, execute as células para verificar os resultados e observar como as alterações afetam a estrutura do grafo.

```python
# Adicionando mais nós e conexões
novas_pessoas = ['Fábio', 'Gabriela', 'Helena']
novas_conexoes = [('Ana', 'Fábio'), ('Bruno', 'Gabriela'), ('Carlos', 'Helena'), ('Fábio', 'Helena'), ('Gabriela', 'Eduardo')]

G.add_nodes_from(novas_pessoas)
G.add_edges_from(novas_conexoes)

# Visualizando novamente
pos = nx.spring_layout(G)  # Pode regenerar posições com novos nós
nx.draw(G, pos, with_labels=True, node_color='lightgreen', edge_color='blue', node_size=1500, font_size=12)
plt.title('Rede Social Ampliada')
plt.show()

# Atualizando informações
print('Novo número de nós:', G.number_of_nodes())
print('Novo número de arestas:', G.number_of_edges())
```

```python
# Parâmetros personalizados de visualização
nx.draw(G, pos, with_labels=True, node_color=['skyblue' if node in pessoas else 'lightgreen' for node in G.nodes()],
        edge_color='purple', node_size=2000, font_size=10)
plt.title('Rede Social com Visualização Personalizada')
plt.show()
```

```python
# Componentes conexos
print('Componentes conexos:')
for i, componente in enumerate(nx.connected_components(G)):
    print(f'Componente {i + 1}:', componente)

# Verificando ciclos
try:
    ciclo = nx.find_cycle(G)
    print('\nCiclo encontrado:', ciclo)
except nx.exception.NetworkXNoCycle:
    print('\nNão há ciclos na rede.')
```

## Seção 5: Extensões e Desafios Adicionais

Para aprofundar seu entendimento, considere os seguintes tópicos:

- **Redes Direcionadas:** Explore como criar e analisar grafos direcionados, onde as arestas possuem uma direção.
- **Redes Ponderadas:** Aprenda a atribuir pesos às arestas para representar a intensidade ou frequência das conexões.
- **Visualização Avançada:** Investigue o uso de outras bibliotecas de visualização, como Plotly, para criar representações interativas.

Você pode implementar esses desafios como exercícios extras, modificando o código base apresentado neste tutorial.

```python
# Criando um grafo direcionado
G_direcionado = nx.DiGraph()
G_direcionado.add_nodes_from(G.nodes)
G_direcionado.add_edges_from(G.edges)

# Visualizando o grafo direcionado
nx.draw(G_direcionado, pos, with_labels=True, node_color='orange', edge_color='black', node_size=1500, font_size=12, arrows=True)
plt.title('Rede Direcionada')
plt.show()
```

```python
# Adicionando pesos às conexões
pesos = {('Ana', 'Bruno'): 5, ('Ana', 'Carlos'): 3, ('Bruno', 'Diana'): 2, ('Carlos', 'Diana'): 4, ('Diana', 'Eduardo'): 1}
nx.set_edge_attributes(G, pesos, 'peso')

# Visualizando pesos no grafo
edge_labels = nx.get_edge_attributes(G, 'peso')
nx.draw(G, pos, with_labels=True, node_color='pink', edge_color='black', node_size=1500, font_size=10)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title('Rede Social Ponderada')
plt.show()
```

```python
import plotly.graph_objects as go

# Configurando a visualização interativa com Plotly
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

node_x = [pos[node][0] for node in G.nodes()]
node_y = [pos[node][1] for node in G.nodes()]
node_text = list(G.nodes())

fig = go.Figure()
fig.add_trace(go.Scatter(x=edge_x, y=edge_y, mode='lines', line=dict(width=1, color='gray')))
fig.add_trace(go.Scatter(x=node_x, y=node_y, mode='markers+text', text=node_text, marker=dict(size=15, color='blue')))
fig.update_layout(title='Rede Social Interativa', showlegend=False)
fig.show()
```
